```python
# Keep only variables we will use in pair plot

sub=df_avo[['Total Volume', 'AveragePrice', 'Small Bags', 'Large Bags','XLarge Bags','Total Bags','region','year','type']]
```

```python
g = sns.pairplot(sub)
```

```python
# Use histogram to visualize distribution of variables. 
# The argument "kde" adds a line that encompasses distribution
sns.histplot(sub['AveragePrice'], bins = 20, kde = True)
#Save figure
plt.savefig('AvgPricehist2.png')
```

```python
sub.loc[sub['AveragePrice'] <= 1.0, 'Average Price Category'] = 'Highest Average Price'
```

```python
sub.loc[(sub['AveragePrice'] > 1.0) & (sub['AveragePrice'] <= 1.5 ), 'Average Price Category'] = 'Average Price'
```

```python
sub.loc[(sub['AveragePrice'] > 1.5) & (sub['AveragePrice'] <= 2 ), 'Average Price Category'] = 'Above Average Price'
```

```python
sub.loc[(sub['AveragePrice'] > 0.5) & (sub['AveragePrice'] <= 1.0 ), 'Average Price Category'] = 'Below Average Price'
```

```python
sub['Average Price Category'].value_counts(dropna = False)
```

```python
sub.columns
```

```python
# Create a categorical plot in seaborn using the categories created above

sns.set(style='ticks')
g = sns.catplot(x='AveragePrice', y='Total Volume', hue='Average Price Category', data=sub)
#Save figure
plt.savefig('AvgPrice_Volume_hist.png')
```
