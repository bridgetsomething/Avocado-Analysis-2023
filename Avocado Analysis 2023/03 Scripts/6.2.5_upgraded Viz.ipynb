```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
%matplotlib inline
```

```python
path = r'/Users/bridget/Downloads/Avocado Analysis 2023'
```

```python
df_avo = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data','avocado_clean.pkl'))
```

```python
#creating a correlation heatmap with matplotlib
plt.matshow(df_avo.corr())
plt.show()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_avo.corr(),cmap='cool', annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
#Save figure
plt.savefig('corfigOG.png')
```

```python
sns.scatterplot(x ='Total Bags', y ='Small Bags', data = df_avo, hue ='year', size = 'type', palette ='YlGnBu' )
#Save figure
plt.savefig('totalbags_smbgs.png')
```

```python
sns.scatterplot(x ='Total Bags', y ='Large Bags', data = df_avo, hue ='year', size = 'type', palette= 'YlOrRd' )
#Save figure
plt.savefig('totalbags_Lgbgs.png')
```

```python
sns.scatterplot(x ='Total Bags', y ='XLarge Bags', data = df_avo, hue ='year', size = 'type', palette='rainbow' )
#Save figure
plt.savefig('totalbags_XLbgs.png')
```

```python
sns.histplot(df_avo['AveragePrice'], kde=True, color='mediumslateblue', bins=20)
plt.savefig('AvgPrcHist.png')
```

```python
sns.displot(df_avo['AveragePrice'], kde=True, color='teal', bins=20)
plt.savefig('AvgPricedisplt.png')
```

```python
sns.barplot(x='type', y='AveragePrice', data= df_avo, palette='summer')
plt.savefig('bar_avgPrc_type.png')
```

```python
sns.jointplot(x='Small Bags', y='AveragePrice', data=df_avo, kind='kde', shade=True, cmap='summer')
plt.savefig('jnt_smBg_avgPr_type.png')
```

```python
sns.jointplot(x='year', y='Total Volume', data=df_avo, kind='kde', shade=True, cmap='cool')
plt.savefig('jnt_TotVol_yr.png')
```

```python
sns.jointplot(x='Small Bags', y='AveragePrice', data=df_avo, kind='hex', cmap='YlGnBu')
plt.savefig('jnt_avgPr_smBg.png')
```

```python
sns.jointplot(x='AveragePrice', y='Total Volume', data=df_avo, kind='hex', cmap='cool')
plt.savefig('jnt_avgPr_TotVol.png')
```

```python
sns.heatmap(df_avo.corr(), annot=True, cmap='icefire')
plt.savefig('heat_cor_ice.png')
```
